```python
# initial_training.py

import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, LayerNormalization, MultiHeadAttention, Add, Input, Reshape, Flatten, BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, accuracy_score, precision_score, recall_score, f1_score
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern
from skimage.measure import label, regionprops
from sklearn.model_selection import KFold
import pywt  # Added for wavelet transformations
import itertools
import joblib  # For saving scaler and label encoder

# Define model save path and naming convention
MODEL_SAVE_PATH = "E:/New folder/00 Temp My Models/"
MODEL_NAME_TEMPLATE = "Model1_{:02d}.h5"  # e.g., Model1_01.h5

# Ensure the directory exists
os.makedirs(MODEL_SAVE_PATH, exist_ok=True)

# Preprocessing functions and feature extraction
def preprocess_image(image_path, target_size=(224, 224)):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image not found or unable to read: {image_path}")
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  
    image = cv2.resize(image, target_size)
    return image

def extract_texture_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    lbp = local_binary_pattern(gray_image, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist / lbp_hist.sum()  
    glcm = graycomatrix(gray_image, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]
    return np.hstack([lbp_hist, contrast, dissimilarity, homogeneity, energy, correlation])

def extract_morphological_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    _, binary_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    labeled_image = label(binary_image)
    regions = regionprops(labeled_image)
    area = [region.area for region in regions]
    eccentricity = [region.eccentricity for region in regions]
    perimeter = [region.perimeter for region in regions]
    if len(area) > 0:
        return np.array([np.mean(area), np.mean(eccentricity), np.mean(perimeter)])
    else:
        return np.array([0, 0, 0])

# New function for wavelet-based feature extraction
def extract_wavelet_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # Perform 2-level wavelet decomposition using 'db1' wavelet
    coeffs = pywt.wavedec2(gray_image, 'db1', level=2)
    features = []
    for coeff in coeffs:
        if isinstance(coeff, tuple):
            # For detailed coefficients at each level
            for arr in coeff:
                features.append(np.mean(arr))
                features.append(np.std(arr))
                features.append(np.max(arr))
                features.append(np.min(arr))
        else:
            # For approximation coefficients
            features.append(np.mean(coeff))
            features.append(np.std(coeff))
            features.append(np.max(coeff))
            features.append(np.min(coeff))
    return np.array(features)

def load_and_split_images(base_path, target_size=(224, 224)):
    images, labels = {'train': [], 'test': [], 'validate': []}, {'train': [], 'test': [], 'validate': []}
    multi_features = {'train': [], 'test': [], 'validate': []}
    for folder_name in os.listdir(base_path):
        class_dir = os.path.join(base_path, folder_name)
        if not os.path.isdir(class_dir):
            continue  # Skip non-directory files
        class_images = [img_name for img_name in os.listdir(class_dir) 
                       if cv2.imread(os.path.join(class_dir, img_name)) is not None]
        if len(class_images) < 1:
            continue  # Skip if no valid images
        np.random.shuffle(class_images)
        total_images = len(class_images)
        train_count = int(total_images * 0.7)
        test_count = int(total_images * 0.15)
        validate_count = total_images - train_count - test_count
        
        def process_images(image_list, label_key, images_dict, multi_features_dict):
            for img_name in image_list:
                img_path = os.path.join(class_dir, img_name)
                try:
                    img = preprocess_image(img_path, target_size)
                except ValueError as e:
                    print(e)
                    continue  # Skip unreadable images
                texture_features = extract_texture_features(img)
                morphological_features = extract_morphological_features(img)
                wavelet_features = extract_wavelet_features(img)  # Extract wavelet features
                # Combine all features
                combined_features = np.hstack([texture_features, morphological_features, wavelet_features])
                multi_features_dict[label_key].append(combined_features)
                images_dict[label_key].append(img)
        
        process_images(class_images[:train_count], 'train', images, multi_features)
        process_images(class_images[train_count:train_count + test_count], 'test', images, multi_features)
        process_images(class_images[train_count + test_count:], 'validate', images, multi_features)
        labels['train'].extend([folder_name] * train_count)
        labels['test'].extend([folder_name] * test_count)
        labels['validate'].extend([folder_name] * validate_count)
    
    for key in images:
        images[key] = np.array(images[key])
        labels[key] = np.array(labels[key])
        multi_features[key] = np.array(multi_features[key])
    return images, labels, multi_features

# Load dataset
base_path = "E:/New folder/Dataset01/batch1"  # Path to the first batch
images, labels, multi_features = load_and_split_images(base_path)

# Normalize images and features
images = {k: preprocess_input(v) for k, v in images.items()}
scaler = StandardScaler()
# Fit on training data and transform all
scaler.fit(multi_features['train'])
multi_features = {k: scaler.transform(v) for k, v in multi_features.items()}

# Encode labels
le = LabelEncoder()
le.fit(labels['train'])  # Fit on training labels
labels_encoded = {k: le.transform(v) for k, v in labels.items()}
labels_cnn = {k: to_categorical(v, num_classes=len(le.classes_)) for k, v in labels_encoded.items()}

# Save the scaler and label encoder
scaler_save_path = os.path.join(MODEL_SAVE_PATH, "scaler.joblib")
le_save_path = os.path.join(MODEL_SAVE_PATH, "label_encoder.joblib")
joblib.dump(scaler, scaler_save_path)
print(f"Scaler saved at: {scaler_save_path}")
joblib.dump(le, le_save_path)
print(f"LabelEncoder saved at: {le_save_path}")

# Extract features from EfficientNetB0
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  

cnn_features_train = base_model.predict(images['train'], verbose=1)
cnn_features_test = base_model.predict(images['test'], verbose=1)
cnn_features_validate = base_model.predict(images['validate'], verbose=1)

def combine_features(cnn_features, multi_features):
    cnn_features_flattened = cnn_features.reshape(cnn_features.shape[0], -1)
    return np.hstack([cnn_features_flattened, multi_features])

X_train_combined = combine_features(cnn_features_train, multi_features['train'])
X_test_combined = combine_features(cnn_features_test, multi_features['test'])
X_validate_combined = combine_features(cnn_features_validate, multi_features['validate'])

# Transformer fusion layer
def transformer_fusion_layer(inputs1, inputs2):
    inputs1 = Reshape((1, -1))(inputs1)
    inputs2 = Reshape((1, -1))(inputs2)
    query = Dense(512)(inputs1)
    key = Dense(512)(inputs2)
    value = Dense(512)(inputs2)
    attention_output = MultiHeadAttention(num_heads=8, key_dim=64)(query, key, value)
    attention_output = Add()([attention_output, query])
    attention_output = LayerNormalization()(attention_output)
    return attention_output

# Build model
input1 = Input(shape=(X_train_combined.shape[1],))
input2 = Input(shape=(multi_features['train'].shape[1],))
fusion_output = transformer_fusion_layer(input1, input2)
fusion_output = BatchNormalization()(fusion_output)
fusion_output = Dropout(0.3)(fusion_output)  
fusion_output = Flatten()(fusion_output)
final_output = Dense(len(le.classes_), activation='softmax')(fusion_output)

model_combined = Model(inputs=[input1, input2], outputs=final_output)

# Learning Rate Scheduler and Model Compilation
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)

model_combined.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Implement cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Arrays to store accuracies and metrics for each fold
train_accuracies = []
val_accuracies = []
test_accuracies = []
fold_metrics = []

fold_no = 1
for train_index, val_index in kf.split(X_train_combined, np.argmax(labels_cnn['train'], axis=1)):
    print(f'Fold {fold_no}')
    
    # Split data into train and validation for each fold
    X_train_fold = X_train_combined[train_index]
    X_val_fold = X_train_combined[val_index]
    y_train_fold = labels_cnn['train'][train_index]
    y_val_fold = labels_cnn['train'][val_index]
    multi_train_fold = multi_features['train'][train_index]
    multi_val_fold = multi_features['train'][val_index]

    # Train the model
    history_combined = model_combined.fit(
        [X_train_fold, multi_train_fold],
        y_train_fold,
        validation_data=([X_val_fold, multi_val_fold], y_val_fold),
        batch_size=32,
        epochs=10,
        callbacks=[reduce_lr]
    )

 
    # Plot accuracy and loss curves
    epochs_range = range(1, len(history_combined.history['accuracy']) + 1)
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, history_combined.history['accuracy'], label='Training Accuracy')
    plt.plot(epochs_range, history_combined.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'Fold {fold_no} - Training and Validation Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, history_combined.history['loss'], label='Training Loss')
    plt.plot(epochs_range, history_combined.history['val_loss'], label='Validation Loss')
    plt.title(f'Fold {fold_no} - Training and Validation Loss')
    plt.legend()
    plt.show()
    
    # Capture train and validation accuracy for each fold
    train_accuracy = history_combined.history['accuracy'][-1]
    val_accuracy = history_combined.history['val_accuracy'][-1]
    
    train_accuracies.append(train_accuracy)
    val_accuracies.append(val_accuracy)

    # Evaluate the model on the test set for the current fold
    y_test_true = np.argmax(labels_cnn['test'], axis=1)
    y_test_pred_prob = model_combined.predict([X_test_combined, multi_features['test']], verbose=1)
    y_test_pred = np.argmax(y_test_pred_prob, axis=1)
    test_accuracy = accuracy_score(y_test_true, y_test_pred)
    test_accuracies.append(test_accuracy)
    
    # Calculate other performance metrics
    precision = precision_score(y_test_true, y_test_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test_true, y_test_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test_true, y_test_pred, average='weighted', zero_division=0)
    
    # Sensitivity and Specificity for each class
    cm = confusion_matrix(y_test_true, y_test_pred)
    sensitivity = cm.diagonal() / cm.sum(axis=1)
    specificity = []
    for i in range(len(le.classes_)):
        tn = cm.sum() - (cm[i, :].sum() + cm[:, i].sum() - cm[i, i])
        fp = cm[:, i].sum() - cm[i, i]
        specificity.append(tn / (tn + fp))
    support = cm.sum(axis=1)
    
    # ROC and AUC
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(len(le.classes_)):
        fpr[i], tpr[i], _ = roc_curve(labels_cnn['test'][:, i], y_test_pred_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Plot ROC curves
    plt.figure()
    colors = itertools.cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red'])
    for i, color in zip(range(len(le.classes_)), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label='ROC curve of class {0} (area = {1:0.4f})'
                 ''.format(le.classes_[i], roc_auc[i]))
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([-0.05, 1.0])
    plt.ylim([0.0, 1.05])
    plt.title(f'Fold {fold_no} - Receiver Operating Characteristic')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
    plt.show()
    
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f'Fold {fold_no} - Confusion Matrix')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    
    # Store metrics
    fold_metrics.append({
        'Fold': fold_no,
        'Train Accuracy': train_accuracy,
        'Validation Accuracy': val_accuracy,
        'Test Accuracy': test_accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Sensitivity': np.mean(sensitivity),
        'Specificity': np.mean(specificity),
        'Support': support.sum()
    })
    
    fold_no += 1

# Compute and display the average accuracies
avg_train_accuracy = np.mean(train_accuracies)
avg_val_accuracy = np.mean(val_accuracies)
avg_test_accuracy = np.mean(test_accuracies)

print(f"Average Training Accuracy across all folds: {avg_train_accuracy*100:.2f}%")
print(f"Average Validation Accuracy across all folds: {avg_val_accuracy*100:.2f}%")
print(f"Average Test Accuracy across all folds: {avg_test_accuracy*100:.2f}%")

# Create a performance table
metrics_df = pd.DataFrame(fold_metrics)
metrics_df['Train Accuracy'] = metrics_df['Train Accuracy'].apply(lambda x: f"{x:.4f}")
metrics_df['Validation Accuracy'] = metrics_df['Validation Accuracy'].apply(lambda x: f"{x:.4f}")
metrics_df['Test Accuracy'] = metrics_df['Test Accuracy'].apply(lambda x: f"{x:.4f}")
metrics_df['Precision'] = metrics_df['Precision'].apply(lambda x: f"{x:.4f}")
metrics_df['Recall'] = metrics_df['Recall'].apply(lambda x: f"{x:.4f}")
metrics_df['F1-Score'] = metrics_df['F1-Score'].apply(lambda x: f"{x:.4f}")
metrics_df['Sensitivity'] = metrics_df['Sensitivity'].apply(lambda x: f"{x:.4f}")
metrics_df['Specificity'] = metrics_df['Specificity'].apply(lambda x: f"{x:.4f}")
metrics_df['Support'] = metrics_df['Support'].astype(int)

print("\nPerformance Metrics for Each Fold:")
print(metrics_df)

# Overall classification report on the test set
y_test_true_all = np.argmax(labels_cnn['test'], axis=1)
y_test_pred_prob_all = model_combined.predict([X_test_combined, multi_features['test']], verbose=1)
y_test_pred_all = np.argmax(y_test_pred_prob_all, axis=1)

report = classification_report(y_test_true_all, y_test_pred_all, target_names=le.classes_, digits=4)
print("\nClassification Report on Test Set:")
print(report)

# Save the trained model
model_save_number = 1  # Starting with 01 for the first model
model_save_name = MODEL_NAME_TEMPLATE.format(model_save_number)
model_save_path = os.path.join(MODEL_SAVE_PATH, model_save_name)

model_combined.save(model_save_path)
print(f"Trained model saved at: {model_save_path}")
```

```python
# incremental_training_with_full_metrics.py

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd  # Added missing import for pandas
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.layers import (
    Dense,
    Dropout,
    LayerNormalization,
    MultiHeadAttention,
    Add,
    Input,
    Reshape,
    Flatten,
    BatchNormalization,
)
from tensorflow.keras.callbacks import ReduceLROnPlateau, Callback  # Added Callback import
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_curve,
    auc,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    cohen_kappa_score,  # Added missing import for cohen_kappa_score
)
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern
from skimage.measure import label, regionprops
from sklearn.model_selection import KFold
import pywt
import itertools
import joblib

# Define model save path and naming convention
MODEL_SAVE_PATH = "E:/New folder/00 Temp My Models/"
MODEL_NAME_TEMPLATE = "Model1_{:02d}.h5"  # e.g., Model1_02.h5

# Preprocessing functions and feature extraction (same as initial training)
def preprocess_image(image_path, target_size=(224, 224)):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image not found or unable to read: {image_path}")
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, target_size)
    return image

def extract_texture_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    lbp = local_binary_pattern(gray_image, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist / lbp_hist.sum()
    glcm = graycomatrix(
        gray_image, distances=[1], angles=[0], levels=256, symmetric=True, normed=True
    )
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]
    return np.hstack(
        [lbp_hist, contrast, dissimilarity, homogeneity, energy, correlation]
    )

def extract_morphological_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    _, binary_image = cv2.threshold(
        gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )
    labeled_image = label(binary_image)
    regions = regionprops(labeled_image)
    area = [region.area for region in regions]
    eccentricity = [region.eccentricity for region in regions]
    perimeter = [region.perimeter for region in regions]
    if len(area) > 0:
        return np.array([np.mean(area), np.mean(eccentricity), np.mean(perimeter)])
    else:
        return np.array([0, 0, 0])

# Function for wavelet-based feature extraction
def extract_wavelet_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # Perform 2-level wavelet decomposition using 'db1' wavelet
    coeffs = pywt.wavedec2(gray_image, 'db1', level=2)
    features = []
    for coeff in coeffs:
        if isinstance(coeff, tuple):
            # For detailed coefficients at each level
            for arr in coeff:
                features.append(np.mean(arr))
                features.append(np.std(arr))
                features.append(np.max(arr))
                features.append(np.min(arr))
        else:
            # For approximation coefficients
            features.append(np.mean(coeff))
            features.append(np.std(coeff))
            features.append(np.max(coeff))
            features.append(np.min(coeff))
    return np.array(features)

def load_and_split_images(base_path, target_size=(224, 224)):
    images, labels = (
        {'train': [], 'test': [], 'validate': []},
        {'train': [], 'test': [], 'validate': []},
    )
    multi_features = {'train': [], 'test': [], 'validate': []}
    for folder_name in os.listdir(base_path):
        class_dir = os.path.join(base_path, folder_name)
        if not os.path.isdir(class_dir):
            continue  # Skip non-directory files
        class_images = [
            img_name
            for img_name in os.listdir(class_dir)
            if cv2.imread(os.path.join(class_dir, img_name)) is not None
        ]
        if len(class_images) < 1:
            continue  # Skip if no valid images
        np.random.shuffle(class_images)
        total_images = len(class_images)
        train_count = int(total_images * 0.7)
        test_count = int(total_images * 0.15)
        validate_count = total_images - train_count - test_count

        def process_images(image_list, label_key, images_dict, multi_features_dict):
            for img_name in image_list:
                img_path = os.path.join(class_dir, img_name)
                try:
                    img = preprocess_image(img_path, target_size)
                except ValueError as e:
                    print(e)
                    continue  # Skip unreadable images
                texture_features = extract_texture_features(img)
                morphological_features = extract_morphological_features(img)
                wavelet_features = extract_wavelet_features(img)  # Extract wavelet features
                # Combine all features
                combined_features = np.hstack(
                    [texture_features, morphological_features, wavelet_features]
                )
                multi_features_dict[label_key].append(combined_features)
                images_dict[label_key].append(img)

        process_images(
            class_images[:train_count], 'train', images, multi_features
        )
        process_images(
            class_images[train_count : train_count + test_count],
            'test',
            images,
            multi_features,
        )
        process_images(
            class_images[train_count + test_count :],
            'validate',
            images,
            multi_features,
        )
        labels['train'].extend([folder_name] * train_count)
        labels['test'].extend([folder_name] * test_count)
        labels['validate'].extend([folder_name] * validate_count)

    for key in images:
        images[key] = np.array(images[key])
        labels[key] = np.array(labels[key])
        multi_features[key] = np.array(multi_features[key])
    return images, labels, multi_features

def combine_features(cnn_features, multi_features):
    cnn_features_flattened = cnn_features.reshape(cnn_features.shape[0], -1)
    return np.hstack([cnn_features_flattened, multi_features])

# Transformer fusion layer (same as initial training)
def transformer_fusion_layer(inputs1, inputs2):
    inputs1 = Reshape((1, -1))(inputs1)
    inputs2 = Reshape((1, -1))(inputs2)
    query = Dense(512)(inputs1)
    key = Dense(512)(inputs2)
    value = Dense(512)(inputs2)
    attention_output = MultiHeadAttention(num_heads=8, key_dim=64)(
        query, key, value
    )
    attention_output = Add()([attention_output, query])
    attention_output = LayerNormalization()(attention_output)
    return attention_output

# Function to save the updated model, scaler, and label encoder
def save_incremental_model(model, model_number, scaler, le):
    model_save_name = MODEL_NAME_TEMPLATE.format(model_number)
    model_save_path = os.path.join(MODEL_SAVE_PATH, model_save_name)
    model.save(model_save_path)
    print(f"Updated model saved at: {model_save_path}")

    # Save the scaler and label encoder
    scaler_save_path = os.path.join(MODEL_SAVE_PATH, f"scaler.joblib")
    le_save_path = os.path.join(MODEL_SAVE_PATH, f"label_encoder.joblib")
    joblib.dump(scaler, scaler_save_path)
    print(f"Scaler saved at: {scaler_save_path}")
    joblib.dump(le, le_save_path)
    print(f"LabelEncoder saved at: {le_save_path}")

# Custom callback to record test metrics
class TestMetricsCallback(Callback):
    def __init__(self, test_data):
        super(TestMetricsCallback, self).__init__()
        self.test_data = test_data
        self.test_accuracies = []
        self.test_losses = []

    def on_epoch_end(self, epoch, logs=None):
        X_test, y_test = self.test_data
        test_loss, test_accuracy = self.model.evaluate(X_test, y_test, verbose=0)
        self.test_accuracies.append(test_accuracy)
        self.test_losses.append(test_loss)
        print(f'Epoch {epoch+1}: Test loss: {test_loss:.4f} - Test accuracy: {test_accuracy:.4f}')

def incremental_train(model_number, batch_path):
    # Load the saved scaler and label encoder
    scaler_load_path = os.path.join(MODEL_SAVE_PATH, "scaler.joblib")
    le_load_path = os.path.join(MODEL_SAVE_PATH, "label_encoder.joblib")

    if not os.path.exists(scaler_load_path):
        raise FileNotFoundError(
            f"Scaler not found at {scaler_load_path}. Ensure it exists."
        )
    if not os.path.exists(le_load_path):
        raise FileNotFoundError(
            f"LabelEncoder not found at {le_load_path}. Ensure it exists."
        )

    scaler = joblib.load(scaler_load_path)
    print(f"Scaler loaded from: {scaler_load_path}")
    le = joblib.load(le_load_path)
    print(f"LabelEncoder loaded from: {le_load_path}")

    # Load the previously saved model without compiling
    previous_model_number = model_number
    previous_model_name = MODEL_NAME_TEMPLATE.format(previous_model_number)
    previous_model_path = os.path.join(MODEL_SAVE_PATH, previous_model_name)

    if not os.path.exists(previous_model_path):
        raise FileNotFoundError(
            f"Previous model not found at {previous_model_path}. Please run the initial training first."
        )

    old_model = load_model(previous_model_path, compile=False)
    print(f"Loaded model from {previous_model_path} without compiling.")

    # Load and process the next batch
    images_new, labels_new, multi_features_new = load_and_split_images(batch_path)

    # Combine new labels with old labels and re-fit the LabelEncoder
    old_classes = le.classes_
    new_classes = np.unique(
        np.concatenate(
            [labels_new['train'], labels_new['test'], labels_new['validate']]
        )
    )
    all_classes = np.unique(np.concatenate([old_classes, new_classes]))
    le = LabelEncoder()
    le.fit(all_classes)
    print(f"Updated classes: {le.classes_}")

    # Encode labels
    labels_encoded_new = {k: le.transform(v) for k, v in labels_new.items()}
    num_classes = len(le.classes_)
    labels_cnn_new = {
        k: to_categorical(v, num_classes=num_classes) for k, v in labels_encoded_new.items()
    }

    # Update the scaler with new data
    scaler.partial_fit(multi_features_new['train'])
    multi_features_new = {
        k: scaler.transform(v) for k, v in multi_features_new.items()
    }

    # Save updated scaler and LabelEncoder
    scaler_save_path = os.path.join(MODEL_SAVE_PATH, "scaler.joblib")
    le_save_path = os.path.join(MODEL_SAVE_PATH, "label_encoder.joblib")
    joblib.dump(scaler, scaler_save_path)
    joblib.dump(le, le_save_path)
    print("Scaler and LabelEncoder updated and saved.")

    # Normalize images
    images_new = {k: preprocess_input(v) for k, v in images_new.items()}

    # Extract features from EfficientNetB0
    base_model = EfficientNetB0(
        weights='imagenet', include_top=False, input_shape=(224, 224, 3)
    )
    base_model.trainable = False

    cnn_features_train_new = base_model.predict(images_new['train'], verbose=1)
    cnn_features_test_new = base_model.predict(images_new['test'], verbose=1)
    cnn_features_validate_new = base_model.predict(images_new['validate'], verbose=1)

    X_train_combined_new = combine_features(
        cnn_features_train_new, multi_features_new['train']
    )
    X_test_combined_new = combine_features(
        cnn_features_test_new, multi_features_new['test']
    )
    X_validate_combined_new = combine_features(
        cnn_features_validate_new, multi_features_new['validate']
    )

    # Rebuild the model to accommodate new classes
    input1 = Input(shape=(X_train_combined_new.shape[1],))
    input2 = Input(shape=(multi_features_new['train'].shape[1],))
    fusion_output = transformer_fusion_layer(input1, input2)
    fusion_output = BatchNormalization()(fusion_output)
    fusion_output = Dropout(0.3)(fusion_output)
    fusion_output = Flatten()(fusion_output)
    final_output = Dense(num_classes, activation='softmax')(fusion_output)
    model_combined = Model(inputs=[input1, input2], outputs=final_output)

    # Transfer weights except for the final layer
    for layer_old, layer_new in zip(old_model.layers[:-1], model_combined.layers[:-1]):
        layer_new.set_weights(layer_old.get_weights())
    print("Weights transferred from the old model to the new model.")

    # Compile the model
    model_combined.compile(
        optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']
    )
    print("Model recompiled with Adam optimizer and categorical_crossentropy loss.")

    # Implement cross-validation
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    # Arrays to store accuracies and metrics for each fold
    train_accuracies = []
    val_accuracies = []
    test_accuracies = []
    fold_metrics = []

    fold_no = 1
    X = X_train_combined_new
    y = np.argmax(labels_cnn_new['train'], axis=1)
    for train_index, val_index in kf.split(X, y):
        print(f'\nFold {fold_no}')

        # Split data into train and validation for each fold
        X_train_fold = X[train_index]
        X_val_fold = X[val_index]
        y_train_fold = labels_cnn_new['train'][train_index]
        y_val_fold = labels_cnn_new['train'][val_index]
        multi_train_fold = multi_features_new['train'][train_index]
        multi_val_fold = multi_features_new['train'][val_index]

        # Create test metrics callback
        test_metrics_callback = TestMetricsCallback(
            test_data=(
                [X_test_combined_new, multi_features_new['test']],
                labels_cnn_new['test'],
            )
        )

        # Train the model
        history_combined = model_combined.fit(
            [X_train_fold, multi_train_fold],
            y_train_fold,
            validation_data=([X_val_fold, multi_val_fold], y_val_fold),
            batch_size=32,
            epochs=10,
            callbacks=[
                ReduceLROnPlateau(
                    monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001
                ),
                test_metrics_callback,
            ],
            verbose=1,
        )

        # Evaluate on training data
        y_train_pred_prob = model_combined.predict(
            [X_train_fold, multi_train_fold], verbose=0
        )
        y_train_pred = np.argmax(y_train_pred_prob, axis=1)
        y_train_true = np.argmax(y_train_fold, axis=1)
        train_accuracy = accuracy_score(y_train_true, y_train_pred)
        train_report = classification_report(
            y_train_true, y_train_pred, target_names=le.classes_, digits=4
        )

        # Evaluate on validation data
        y_val_pred_prob = model_combined.predict(
            [X_val_fold, multi_val_fold], verbose=0
        )
        y_val_pred = np.argmax(y_val_pred_prob, axis=1)
        y_val_true = np.argmax(y_val_fold, axis=1)
        val_accuracy = accuracy_score(y_val_true, y_val_pred)
        val_report = classification_report(
            y_val_true, y_val_pred, target_names=le.classes_, digits=4
        )

        # Evaluate on test data
        y_test_true = np.argmax(labels_cnn_new['test'], axis=1)
        y_test_pred_prob = model_combined.predict(
            [X_test_combined_new, multi_features_new['test']], verbose=0
        )
        y_test_pred = np.argmax(y_test_pred_prob, axis=1)
        test_accuracy = accuracy_score(y_test_true, y_test_pred)
        test_report = classification_report(
            y_test_true, y_test_pred, target_names=le.classes_, digits=4
        )

        # Calculate Kappa Value
        kappa = cohen_kappa_score(y_test_true, y_test_pred)
        print(f"Cohen's Kappa Score: {kappa:.4f}")

        # Append accuracies
        train_accuracies.append(train_accuracy)
        val_accuracies.append(val_accuracy)
        test_accuracies.append(test_accuracy)

        # Store metrics
        fold_metrics.append(
            {
                'Fold': fold_no,
                'Train Accuracy': train_accuracy,
                'Validation Accuracy': val_accuracy,
                'Test Accuracy': test_accuracy,
                'Kappa': kappa,
            }
        )

        # Plot Accuracy and Loss Curves
        epochs_range = range(1, len(history_combined.history['accuracy']) + 1)
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(
            epochs_range,
            history_combined.history['accuracy'],
            label='Training Accuracy',
        )
        plt.plot(
            epochs_range,
            history_combined.history['val_accuracy'],
            label='Validation Accuracy',
        )
        plt.plot(
            epochs_range,
            test_metrics_callback.test_accuracies,
            label='Testing Accuracy',
        )
        plt.title(f'Fold {fold_no} - Accuracy')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(
            epochs_range, history_combined.history['loss'], label='Training Loss'
        )
        plt.plot(
            epochs_range,
            history_combined.history['val_loss'],
            label='Validation Loss',
        )
        plt.plot(
            epochs_range, test_metrics_callback.test_losses, label='Testing Loss'
        )
        plt.title(f'Fold {fold_no} - Loss')
        plt.legend()
        plt.show()

        # ROC and AUC for Test Set
        fpr_test = dict()
        tpr_test = dict()
        roc_auc_test = dict()
        for i in range(num_classes):
            fpr_test[i], tpr_test[i], _ = roc_curve(
                labels_cnn_new['test'][:, i], y_test_pred_prob[:, i]
            )
            roc_auc_test[i] = auc(fpr_test[i], tpr_test[i])

        # Plot ROC curves for Test Set
        plt.figure()
        colors = itertools.cycle(
            ['aqua', 'darkorange', 'cornflowerblue', 'green', 'red']
        )
        for i, color in zip(range(num_classes), colors):
            plt.plot(
                fpr_test[i],
                tpr_test[i],
                color=color,
                lw=2,
                label='ROC curve of class {0} (area = {1:0.4f})'.format(
                    le.classes_[i], roc_auc_test[i]
                ),
            )
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([-0.05, 1.0])
        plt.ylim([0.0, 1.05])
        plt.title(f'Fold {fold_no} - ROC Curve (Test Set)')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend(loc="lower right")
        plt.show()

        # Confusion Matrix for Test Set
        cm_test = confusion_matrix(y_test_true, y_test_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(
            cm_test,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_,
        )
        plt.title(f'Fold {fold_no} - Confusion Matrix (Test Set)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.show()

        # Print classification reports
        print("\nClassification Report - Training Set:")
        print(train_report)
        print("\nClassification Report - Validation Set:")
        print(val_report)
        print("\nClassification Report - Test Set:")
        print(test_report)

        fold_no += 1

    # Compute and display the average accuracies
    avg_train_accuracy = np.mean(train_accuracies)
    avg_val_accuracy = np.mean(val_accuracies)
    avg_test_accuracy = np.mean(test_accuracies)

    print(f"\nAverage Training Accuracy across all folds: {avg_train_accuracy*100:.2f}%")
    print(f"Average Validation Accuracy across all folds: {avg_val_accuracy*100:.2f}%")
    print(f"Average Test Accuracy across all folds: {avg_test_accuracy*100:.2f}%")

    # Create a performance table
    metrics_df = pd.DataFrame(fold_metrics)
    metrics_df['Train Accuracy'] = metrics_df['Train Accuracy'].apply(lambda x: f"{x:.4f}")
    metrics_df['Validation Accuracy'] = metrics_df['Validation Accuracy'].apply(lambda x: f"{x:.4f}")
    metrics_df['Test Accuracy'] = metrics_df['Test Accuracy'].apply(lambda x: f"{x:.4f}")
    metrics_df['Kappa'] = metrics_df['Kappa'].apply(lambda x: f"{x:.4f}")

    print("\nPerformance Metrics for Each Fold:")
    print(metrics_df)

    # Overall classification report on the test set
    y_test_true_all = np.argmax(labels_cnn_new['test'], axis=1)
    y_test_pred_prob_all = model_combined.predict(
        [X_test_combined_new, multi_features_new['test']], verbose=1
    )
    y_test_pred_all = np.argmax(y_test_pred_prob_all, axis=1)
    overall_test_accuracy = accuracy_score(y_test_true_all, y_test_pred_all)
    overall_kappa = cohen_kappa_score(y_test_true_all, y_test_pred_all)
    overall_test_report = classification_report(
        y_test_true_all, y_test_pred_all, target_names=le.classes_, digits=4
    )

    print(f"\nOverall Test Accuracy: {overall_test_accuracy*100:.2f}%")
    print(f"Overall Cohen's Kappa Score: {overall_kappa:.4f}")
    print("\nOverall Classification Report on Test Set:")
    print(overall_test_report)

    # Save the updated model
    new_model_number = model_number + 1
    save_incremental_model(model_combined, new_model_number, scaler, le)

if __name__ == "__main__":
    # Define the next batch path (e.g., batch3)
    next_batch_path = "E:/New folder/Dataset01/batch2"  # Update this path to your next batch

    # Current model number (Model1_02.h5)
    current_model_number = 3  # Set to the model number corresponding to your batch 2 trained model

    # Perform incremental training
    incremental_train(current_model_number, next_batch_path)
```

```python
# incremental_training_with_full_metrics.py

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd  # Added missing import for pandas
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.layers import (
    Dense,
    Dropout,
    LayerNormalization,
    MultiHeadAttention,
    Add,
    Input,
    Reshape,
    Flatten,
    BatchNormalization,
)
from tensorflow.keras.callbacks import ReduceLROnPlateau, Callback  # Added Callback import
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_curve,
    auc,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    cohen_kappa_score,  # Added missing import for cohen_kappa_score
)
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern
from skimage.measure import label, regionprops
from sklearn.model_selection import KFold
import pywt
import itertools
import joblib

# Define model save path and naming convention
MODEL_SAVE_PATH = "E:/New folder/00 Temp My Models/"
MODEL_NAME_TEMPLATE = "Model1_{:02d}.h5"  # e.g., Model1_02.h5

# Preprocessing functions and feature extraction (same as initial training)
def preprocess_image(image_path, target_size=(224, 224)):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image not found or unable to read: {image_path}")
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, target_size)
    return image

def extract_texture_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    lbp = local_binary_pattern(gray_image, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), range=(0, 10))
    lbp_hist = lbp_hist / lbp_hist.sum()
    glcm = graycomatrix(
        gray_image, distances=[1], angles=[0], levels=256, symmetric=True, normed=True
    )
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]
    return np.hstack(
        [lbp_hist, contrast, dissimilarity, homogeneity, energy, correlation]
    )

def extract_morphological_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    _, binary_image = cv2.threshold(
        gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )
    labeled_image = label(binary_image)
    regions = regionprops(labeled_image)
    area = [region.area for region in regions]
    eccentricity = [region.eccentricity for region in regions]
    perimeter = [region.perimeter for region in regions]
    if len(area) > 0:
        return np.array([np.mean(area), np.mean(eccentricity), np.mean(perimeter)])
    else:
        return np.array([0, 0, 0])

# Function for wavelet-based feature extraction
def extract_wavelet_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # Perform 2-level wavelet decomposition using 'db1' wavelet
    coeffs = pywt.wavedec2(gray_image, 'db1', level=2)
    features = []
    for coeff in coeffs:
        if isinstance(coeff, tuple):
            # For detailed coefficients at each level
            for arr in coeff:
                features.append(np.mean(arr))
                features.append(np.std(arr))
                features.append(np.max(arr))
                features.append(np.min(arr))
        else:
            # For approximation coefficients
            features.append(np.mean(coeff))
            features.append(np.std(coeff))
            features.append(np.max(coeff))
            features.append(np.min(coeff))
    return np.array(features)

def load_and_split_images(base_path, target_size=(224, 224)):
    images, labels = (
        {'train': [], 'test': [], 'validate': []},
        {'train': [], 'test': [], 'validate': []},
    )
    multi_features = {'train': [], 'test': [], 'validate': []}
    for folder_name in os.listdir(base_path):
        class_dir = os.path.join(base_path, folder_name)
        if not os.path.isdir(class_dir):
            continue  # Skip non-directory files
        class_images = [
            img_name
            for img_name in os.listdir(class_dir)
            if cv2.imread(os.path.join(class_dir, img_name)) is not None
        ]
        if len(class_images) < 1:
            continue  # Skip if no valid images
        np.random.shuffle(class_images)
        total_images = len(class_images)
        train_count = int(total_images * 0.7)
        test_count = int(total_images * 0.15)
        validate_count = total_images - train_count - test_count

        def process_images(image_list, label_key, images_dict, multi_features_dict):
            for img_name in image_list:
                img_path = os.path.join(class_dir, img_name)
                try:
                    img = preprocess_image(img_path, target_size)
                except ValueError as e:
                    print(e)
                    continue  # Skip unreadable images
                texture_features = extract_texture_features(img)
                morphological_features = extract_morphological_features(img)
                wavelet_features = extract_wavelet_features(img)  # Extract wavelet features
                # Combine all features
                combined_features = np.hstack(
                    [texture_features, morphological_features, wavelet_features]
                )
                multi_features_dict[label_key].append(combined_features)
                images_dict[label_key].append(img)

        process_images(
            class_images[:train_count], 'train', images, multi_features
        )
        process_images(
            class_images[train_count : train_count + test_count],
            'test',
            images,
            multi_features,
        )
        process_images(
            class_images[train_count + test_count :],
            'validate',
            images,
            multi_features,
        )
        labels['train'].extend([folder_name] * train_count)
        labels['test'].extend([folder_name] * test_count)
        labels['validate'].extend([folder_name] * validate_count)

    for key in images:
        images[key] = np.array(images[key])
        labels[key] = np.array(labels[key])
        multi_features[key] = np.array(multi_features[key])
    return images, labels, multi_features

def combine_features(cnn_features, multi_features):
    cnn_features_flattened = cnn_features.reshape(cnn_features.shape[0], -1)
    return np.hstack([cnn_features_flattened, multi_features])

# Transformer fusion layer (same as initial training)
def transformer_fusion_layer(inputs1, inputs2):
    inputs1 = Reshape((1, -1))(inputs1)
    inputs2 = Reshape((1, -1))(inputs2)
    query = Dense(512)(inputs1)
    key = Dense(512)(inputs2)
    value = Dense(512)(inputs2)
    attention_output = MultiHeadAttention(num_heads=8, key_dim=64)(
        query, key, value
    )
    attention_output = Add()([attention_output, query])
    attention_output = LayerNormalization()(attention_output)
    return attention_output

# Function to save the updated model, scaler, and label encoder
def save_incremental_model(model, model_number, scaler, le):
    model_save_name = MODEL_NAME_TEMPLATE.format(model_number)
    model_save_path = os.path.join(MODEL_SAVE_PATH, model_save_name)
    model.save(model_save_path)
    print(f"Updated model saved at: {model_save_path}")

    # Save the scaler and label encoder
    scaler_save_path = os.path.join(MODEL_SAVE_PATH, f"scaler.joblib")
    le_save_path = os.path.join(MODEL_SAVE_PATH, f"label_encoder.joblib")
    joblib.dump(scaler, scaler_save_path)
    print(f"Scaler saved at: {scaler_save_path}")
    joblib.dump(le, le_save_path)
    print(f"LabelEncoder saved at: {le_save_path}")

# Custom callback to record test metrics
class TestMetricsCallback(Callback):
    def __init__(self, test_data):
        super(TestMetricsCallback, self).__init__()
        self.test_data = test_data
        self.test_accuracies = []
        self.test_losses = []

    def on_epoch_end(self, epoch, logs=None):
        X_test, y_test = self.test_data
        test_loss, test_accuracy = self.model.evaluate(X_test, y_test, verbose=0)
        self.test_accuracies.append(test_accuracy)
        self.test_losses.append(test_loss)
        print(f'Epoch {epoch+1}: Test loss: {test_loss:.4f} - Test accuracy: {test_accuracy:.4f}')

def incremental_train(model_number, batch_path):
    # Load the saved scaler and label encoder
    scaler_load_path = os.path.join(MODEL_SAVE_PATH, "scaler.joblib")
    le_load_path = os.path.join(MODEL_SAVE_PATH, "label_encoder.joblib")

    if not os.path.exists(scaler_load_path):
        raise FileNotFoundError(
            f"Scaler not found at {scaler_load_path}. Ensure it exists."
        )
    if not os.path.exists(le_load_path):
        raise FileNotFoundError(
            f"LabelEncoder not found at {le_load_path}. Ensure it exists."
        )

    scaler = joblib.load(scaler_load_path)
    print(f"Scaler loaded from: {scaler_load_path}")
    le = joblib.load(le_load_path)
    print(f"LabelEncoder loaded from: {le_load_path}")

    # Load the previously saved model without compiling
    previous_model_number = model_number
    previous_model_name = MODEL_NAME_TEMPLATE.format(previous_model_number)
    previous_model_path = os.path.join(MODEL_SAVE_PATH, previous_model_name)

    if not os.path.exists(previous_model_path):
        raise FileNotFoundError(
            f"Previous model not found at {previous_model_path}. Please run the initial training first."
        )

    old_model = load_model(previous_model_path, compile=False)
    print(f"Loaded model from {previous_model_path} without compiling.")

    # Load and process the next batch
    images_new, labels_new, multi_features_new = load_and_split_images(batch_path)

    # Combine new labels with old labels and re-fit the LabelEncoder
    old_classes = le.classes_
    new_classes = np.unique(
        np.concatenate(
            [labels_new['train'], labels_new['test'], labels_new['validate']]
        )
    )
    all_classes = np.unique(np.concatenate([old_classes, new_classes]))
    le = LabelEncoder()
    le.fit(all_classes)
    print(f"Updated classes: {le.classes_}")

    # Encode labels
    labels_encoded_new = {k: le.transform(v) for k, v in labels_new.items()}
    num_classes = len(le.classes_)
    labels_cnn_new = {
        k: to_categorical(v, num_classes=num_classes) for k, v in labels_encoded_new.items()
    }

    # Update the scaler with new data
    scaler.partial_fit(multi_features_new['train'])
    multi_features_new = {
        k: scaler.transform(v) for k, v in multi_features_new.items()
    }

    # Save updated scaler and LabelEncoder
    scaler_save_path = os.path.join(MODEL_SAVE_PATH, "scaler.joblib")
    le_save_path = os.path.join(MODEL_SAVE_PATH, "label_encoder.joblib")
    joblib.dump(scaler, scaler_save_path)
    joblib.dump(le, le_save_path)
    print("Scaler and LabelEncoder updated and saved.")

    # Normalize images
    images_new = {k: preprocess_input(v) for k, v in images_new.items()}

    # Extract features from EfficientNetB0
    base_model = EfficientNetB0(
        weights='imagenet', include_top=False, input_shape=(224, 224, 3)
    )
    base_model.trainable = False

    cnn_features_train_new = base_model.predict(images_new['train'], verbose=1)
    cnn_features_test_new = base_model.predict(images_new['test'], verbose=1)
    cnn_features_validate_new = base_model.predict(images_new['validate'], verbose=1)

    X_train_combined_new = combine_features(
        cnn_features_train_new, multi_features_new['train']
    )
    X_test_combined_new = combine_features(
        cnn_features_test_new, multi_features_new['test']
    )
    X_validate_combined_new = combine_features(
        cnn_features_validate_new, multi_features_new['validate']
    )

    # Rebuild the model to accommodate new classes
    input1 = Input(shape=(X_train_combined_new.shape[1],))
    input2 = Input(shape=(multi_features_new['train'].shape[1],))
    fusion_output = transformer_fusion_layer(input1, input2)
    fusion_output = BatchNormalization()(fusion_output)
    fusion_output = Dropout(0.3)(fusion_output)
    fusion_output = Flatten()(fusion_output)
    final_output = Dense(num_classes, activation='softmax')(fusion_output)
    model_combined = Model(inputs=[input1, input2], outputs=final_output)

    # Transfer weights except for the final layer
    for layer_old, layer_new in zip(old_model.layers[:-1], model_combined.layers[:-1]):
        layer_new.set_weights(layer_old.get_weights())
    print("Weights transferred from the old model to the new model.")

    # Compile the model
    model_combined.compile(
        optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']
    )
    print("Model recompiled with Adam optimizer and categorical_crossentropy loss.")

    # Implement cross-validation
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    # Arrays to store accuracies and metrics for each fold
    train_accuracies = []
    val_accuracies = []
    test_accuracies = []
    fold_metrics = []

    fold_no = 1
    X = X_train_combined_new
    y = np.argmax(labels_cnn_new['train'], axis=1)
    for train_index, val_index in kf.split(X, y):
        print(f'\nFold {fold_no}')

        # Split data into train and validation for each fold
        X_train_fold = X[train_index]
        X_val_fold = X[val_index]
        y_train_fold = labels_cnn_new['train'][train_index]
        y_val_fold = labels_cnn_new['train'][val_index]
        multi_train_fold = multi_features_new['train'][train_index]
        multi_val_fold = multi_features_new['train'][val_index]

        # Create test metrics callback
        test_metrics_callback = TestMetricsCallback(
            test_data=(
                [X_test_combined_new, multi_features_new['test']],
                labels_cnn_new['test'],
            )
        )

        # Train the model
        history_combined = model_combined.fit(
            [X_train_fold, multi_train_fold],
            y_train_fold,
            validation_data=([X_val_fold, multi_val_fold], y_val_fold),
            batch_size=32,
            epochs=10,
            callbacks=[
                ReduceLROnPlateau(
                    monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001
                ),
                test_metrics_callback,
            ],
            verbose=1,
        )

        # Evaluate on training data
        y_train_pred_prob = model_combined.predict(
            [X_train_fold, multi_train_fold], verbose=0
        )
        y_train_pred = np.argmax(y_train_pred_prob, axis=1)
        y_train_true = np.argmax(y_train_fold, axis=1)
        train_accuracy = accuracy_score(y_train_true, y_train_pred)
        train_report = classification_report(
            y_train_true, y_train_pred, target_names=le.classes_, digits=4
        )

        # Evaluate on validation data
        y_val_pred_prob = model_combined.predict(
            [X_val_fold, multi_val_fold], verbose=0
        )
        y_val_pred = np.argmax(y_val_pred_prob, axis=1)
        y_val_true = np.argmax(y_val_fold, axis=1)
        val_accuracy = accuracy_score(y_val_true, y_val_pred)
        val_report = classification_report(
            y_val_true, y_val_pred, target_names=le.classes_, digits=4
        )

        # Evaluate on test data
        y_test_true = np.argmax(labels_cnn_new['test'], axis=1)
        y_test_pred_prob = model_combined.predict(
            [X_test_combined_new, multi_features_new['test']], verbose=0
        )
        y_test_pred = np.argmax(y_test_pred_prob, axis=1)
        test_accuracy = accuracy_score(y_test_true, y_test_pred)
        test_report = classification_report(
            y_test_true, y_test_pred, target_names=le.classes_, digits=4
        )

        # Calculate Kappa Value
        kappa = cohen_kappa_score(y_test_true, y_test_pred)
        print(f"Cohen's Kappa Score: {kappa:.4f}")

        # Append accuracies
        train_accuracies.append(train_accuracy)
        val_accuracies.append(val_accuracy)
        test_accuracies.append(test_accuracy)

        # Store metrics
        fold_metrics.append(
            {
                'Fold': fold_no,
                'Train Accuracy': train_accuracy,
                'Validation Accuracy': val_accuracy,
                'Test Accuracy': test_accuracy,
                'Kappa': kappa,
            }
        )

        # Plot Accuracy and Loss Curves
        epochs_range = range(1, len(history_combined.history['accuracy']) + 1)
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(
            epochs_range,
            history_combined.history['accuracy'],
            label='Training Accuracy',
        )
        plt.plot(
            epochs_range,
            history_combined.history['val_accuracy'],
            label='Validation Accuracy',
        )
        plt.plot(
            epochs_range,
            test_metrics_callback.test_accuracies,
            label='Testing Accuracy',
        )
        plt.title(f'Fold {fold_no} - Accuracy')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(
            epochs_range, history_combined.history['loss'], label='Training Loss'
        )
        plt.plot(
            epochs_range,
            history_combined.history['val_loss'],
            label='Validation Loss',
        )
        plt.plot(
            epochs_range, test_metrics_callback.test_losses, label='Testing Loss'
        )
        plt.title(f'Fold {fold_no} - Loss')
        plt.legend()
        plt.show()

        # ROC and AUC for Test Set
        fpr_test = dict()
        tpr_test = dict()
        roc_auc_test = dict()
        for i in range(num_classes):
            fpr_test[i], tpr_test[i], _ = roc_curve(
                labels_cnn_new['test'][:, i], y_test_pred_prob[:, i]
            )
            roc_auc_test[i] = auc(fpr_test[i], tpr_test[i])

        # Plot ROC curves for Test Set
        plt.figure()
        colors = itertools.cycle(
            ['aqua', 'darkorange', 'cornflowerblue', 'green', 'red']
        )
        for i, color in zip(range(num_classes), colors):
            plt.plot(
                fpr_test[i],
                tpr_test[i],
                color=color,
                lw=2,
                label='ROC curve of class {0} (area = {1:0.4f})'.format(
                    le.classes_[i], roc_auc_test[i]
                ),
            )
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([-0.05, 1.0])
        plt.ylim([0.0, 1.05])
        plt.title(f'Fold {fold_no} - ROC Curve (Test Set)')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend(loc="lower right")
        plt.show()

        # Confusion Matrix for Test Set
        cm_test = confusion_matrix(y_test_true, y_test_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(
            cm_test,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_,
        )
        plt.title(f'Fold {fold_no} - Confusion Matrix (Test Set)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.show()

        # Print classification reports
        print("\nClassification Report - Training Set:")
        print(train_report)
        print("\nClassification Report - Validation Set:")
        print(val_report)
        print("\nClassification Report - Test Set:")
        print(test_report)

        fold_no += 1

    # Compute and display the average accuracies
    avg_train_accuracy = np.mean(train_accuracies)
    avg_val_accuracy = np.mean(val_accuracies)
    avg_test_accuracy = np.mean(test_accuracies)

    print(f"\nAverage Training Accuracy across all folds: {avg_train_accuracy*100:.2f}%")
    print(f"Average Validation Accuracy across all folds: {avg_val_accuracy*100:.2f}%")
    print(f"Average Test Accuracy across all folds: {avg_test_accuracy*100:.2f}%")

    # Create a performance table
    metrics_df = pd.DataFrame(fold_metrics)
    metrics_df['Train Accuracy'] = metrics_df['Train Accuracy'].apply(lambda x: f"{x:.4f}")
    metrics_df['Validation Accuracy'] = metrics_df['Validation Accuracy'].apply(lambda x: f"{x:.4f}")
    metrics_df['Test Accuracy'] = metrics_df['Test Accuracy'].apply(lambda x: f"{x:.4f}")
    metrics_df['Kappa'] = metrics_df['Kappa'].apply(lambda x: f"{x:.4f}")

    print("\nPerformance Metrics for Each Fold:")
    print(metrics_df)

    # Overall classification report on the test set
    y_test_true_all = np.argmax(labels_cnn_new['test'], axis=1)
    y_test_pred_prob_all = model_combined.predict(
        [X_test_combined_new, multi_features_new['test']], verbose=1
    )
    y_test_pred_all = np.argmax(y_test_pred_prob_all, axis=1)
    overall_test_accuracy = accuracy_score(y_test_true_all, y_test_pred_all)
    overall_kappa = cohen_kappa_score(y_test_true_all, y_test_pred_all)
    overall_test_report = classification_report(
        y_test_true_all, y_test_pred_all, target_names=le.classes_, digits=4
    )

    print(f"\nOverall Test Accuracy: {overall_test_accuracy*100:.2f}%")
    print(f"Overall Cohen's Kappa Score: {overall_kappa:.4f}")
    print("\nOverall Classification Report on Test Set:")
    print(overall_test_report)

    # Save the updated model
    new_model_number = model_number + 1
    save_incremental_model(model_combined, new_model_number, scaler, le)

if __name__ == "__main__":
    # Define the next batch path (e.g., batch3)
    next_batch_path = "E:/New folder/Dataset01/batch4"  # Update this path to your next batch

    # Current model number (Model1_02.h5)
    current_model_number =4  # Set to the model number corresponding to your batch 2 trained model

    # Perform incremental training
    incremental_train(current_model_number, next_batch_path)
```
